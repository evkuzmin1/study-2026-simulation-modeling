# Динамика числа маргариток при разных параметрах модели

В данной работе исследуется изменение числа чёрных и белых ромашек
в зависимости от модельного времени при различных параметрах модели Daisyworld.

В качестве изменяемых параметров рассматриваются:
- максимальный возраст ромашек (`max_age`);
- начальная доля белых ромашек (`init_white`).

Для каждой комбинации параметров строится отдельный график.

```julia
using DrWatson
@quickactivate "project"
```

## Подключение библиотек

```julia
using Agents
using DataFrames
using Plots
using CairoMakie
```

## Подключение модели

```julia
include(srcdir("daisyworld.jl"))
```

## Определение функций для подсчёта агентов

```julia
black(a) = a.breed == :black
white(a) = a.breed == :white
adata = [(black, count), (white, count)]
```

## Параметры эксперимента

```julia
param_dict = Dict(
    :griddims => (30, 30),
    :max_age => [25, 40],
    :init_white => [0.2, 0.8],
    :init_black => 0.2,
    :albedo_white => 0.75,
    :albedo_black => 0.25,
    :surface_albedo => 0.4,
    :solar_change => 0.005,
    :solar_luminosity => 1.0,
    :scenario => :default,
    :seed => 165,
)
```

## Формирование списка комбинаций параметров

```julia
params_list = dict_list(param_dict)
```

## Проведение серии экспериментов

```julia
for params in params_list
    model = daisyworld(; params...)
```

Запуск моделирования на 1000 шагов

```julia
    agent_df, model_df = run!(model, 1000; adata)
```

## Построение графика численности ромашек

```julia
    figure = Figure(size = (600, 400))

    ax = figure[1, 1] = Axis(
        figure,
        xlabel = "tick",
        ylabel = "daisy count"
    )

    blackl = lines!(
        ax,
        agent_df[!, :time],
        agent_df[!, :count_black],
        color = :black
    )

    whitel = lines!(
        ax,
        agent_df[!, :time],
        agent_df[!, :count_white],
        color = :orange
    )

    Legend(
        figure[1, 2],
        [blackl, whitel],
        ["black", "white"],
        labelsize = 12
    )
```

## Сохранение результатов

```julia
    plt_name = savename("daisy-count", params) * ".png"
    save(plotsdir(plt_name), figure)
end

println("Готово: графики с параметрами сохранены в папку plots")
```

## Анализ результатов

По полученным графикам можно сделать следующие выводы:

1. На начальном этапе для разных комбинаций параметров наблюдаются
   резкие колебания численности чёрных и белых ромашек.

2. По мере развития моделирования система выходит на устойчивый режим,
   при котором численности обеих популяций стабилизируются в определённых пределах.

3. Линии численности чёрных и белых ромашек не совпадают, поскольку
   данные типы агентов выполняют разные функции в модели:
   чёрные ромашки нагревают поверхность, а белые — охлаждают её.

4. Изменение параметров влияет на форму колебаний и итоговый баланс
   между популяциями, однако общая тенденция к саморегуляции сохраняется.

Таким образом, модель Daisyworld демонстрирует устойчивую динамику
численности популяций даже при варьировании параметров.
