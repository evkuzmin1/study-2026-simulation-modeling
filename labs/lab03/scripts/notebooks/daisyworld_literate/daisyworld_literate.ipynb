# Базовая визуализация модели Daisyworld

В данной работе рассматривается агентная модель Daisyworld.
Целью является построение базовой визуализации модели, в которой:
- цвет фона отражает температуру поверхности;
- белые и чёрные ромашки отображаются в соответствии со своим типом;
- анализируется состояние системы на нескольких шагах моделирования.

Сначала подключаются необходимые библиотеки и активируется окружение проекта.

```julia
using DrWatson
@quickactivate
```

## Подключение библиотек

```julia
using Agents
using DataFrames
using Plots
using CairoMakie
```

## Подключение файла модели

Основная логика модели находится в файле `src/daisyworld.jl`.

```julia
include(srcdir("daisyworld.jl"))
```

## Создание модели

Создаём экземпляр модели Daisyworld с параметрами по умолчанию.

```julia
model = daisyworld()
```

## Настройка визуализации

Для отображения ромашек используется их тип:
- белые ромашки;
- чёрные ромашки.

Температура отображается как тепловая карта.

```julia
daisycolor(a::Daisy) = a.breed

plotkwargs = (
    agent_color = daisycolor,
    agent_size = 20,
    agent_marker = '✿',
    heatarray = :temperature,
    heatkwargs = (colorrange = (-20, 60),),
)
```

## Построение первого состояния

На первом шаге система находится в начальном состоянии.
Распределение ромашек является случайным.

```julia
plt1, _ = abmplot(model; plotkwargs...)
```

## Построение состояния после 5 шагов

После нескольких шагов моделирования начинается формирование
локальных температурных зон.

```julia
step!(model, 5)
plt2, _ = abmplot(model; heatarray = model.temperature, plotkwargs...)
```

## Построение состояния после 40 шагов

На более позднем этапе система демонстрирует самоорганизацию:
распределение ромашек влияет на температурный режим среды.

```julia
step!(model, 40)
plt3, _ = abmplot(model; heatarray = model.temperature, plotkwargs...)
```

## Сохранение результатов

Полученные изображения сохраняются в папку `plots`.

```julia
save(plotsdir("daisy_step001.png"), plt1)
save(plotsdir("daisy_step005.png"), plt2)
save(plotsdir("daisy_step040.png"), plt3)

println("Готово: сохранены файлы daisy_step001.png, daisy_step005.png, daisy_step040.png")
```

## Анализ результатов

По полученным изображениям можно сделать следующие выводы:

1. В начальный момент распределение ромашек является случайным,
   а температурное поле ещё не демонстрирует выраженной структуры.

2. На ранних этапах моделирования начинают формироваться локальные зоны
   с различной температурой, что связано с различием альбедо ромашек.

3. На более позднем этапе видно, что система приходит к самоорганизации:
   распределение белых и чёрных ромашек влияет на температуру поверхности
   и способствует её саморегуляции.

Таким образом, модель Daisyworld демонстрирует механизм саморегуляции
среды за счёт взаимодействия агентов и окружающей среды.
