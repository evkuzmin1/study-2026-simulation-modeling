# Динамика числа маргариток в модели Daisyworld

В данной работе строится график изменения числа чёрных и белых ромашек
в зависимости от модельного времени.

Цель работы — проследить, как изменяется численность популяций во времени
и сделать вывод о характере динамики системы.

```julia
using DrWatson
@quickactivate "project"
```

## Подключение библиотек

```julia
using Agents
using DataFrames
using Plots
using CairoMakie
```

## Подключение модели

```julia
include(srcdir("daisyworld.jl"))
```

## Определение функций для подсчёта ромашек

Функции `black` и `white` позволяют определить, к какому типу
относится агент, а `adata` задаёт агрегируемые данные для подсчёта
количества агентов каждого типа.

```julia
black(a) = a.breed == :black
white(a) = a.breed == :white
adata = [(black, count), (white, count)]
```

## Создание модели

Создаётся модель Daisyworld с начальной светимостью 1.0.

```julia
model = daisyworld(; solar_luminosity = 1.0)
```

## Запуск моделирования

Выполняется моделирование на 1000 шагов.
Результаты по агентам сохраняются в `agent_df`.

```julia
agent_df, model_df = run!(model, 1000; adata)
```

## Построение графика

На графике отображаются:
- число чёрных ромашек;
- число белых ромашек.

```julia
figure = Figure(size = (600, 400))

ax = figure[1, 1] = Axis(
    figure,
    xlabel = "tick",
    ylabel = "daisy count"
)

blackl = lines!(
    ax,
    agent_df[!, :time],
    agent_df[!, :count_black],
    color = :black
)

whitel = lines!(
    ax,
    agent_df[!, :time],
    agent_df[!, :count_white],
    color = :orange
)

Legend(
    figure[1, 2],
    [blackl, whitel],
    ["black", "white"],
    labelsize = 12
)
```

## Сохранение результата

График сохраняется в файл `daisy_count.png` в папке `plots`.

```julia
save(plotsdir("daisy_count.png"), figure)

println("Готово: создан файл plots/daisy_count.png")
```

## Анализ результата

По полученному графику можно сделать следующие выводы:

1. На начальном этапе наблюдаются резкие колебания численности
   чёрных и белых ромашек, что связано со случайным распределением
   агентов и отсутствием сформированного температурного режима.

2. После начального периода система переходит в более устойчивое
   состояние, при котором численность обеих популяций изменяется
   в ограниченном диапазоне.

3. Численности чёрных и белых ромашек колеблются согласованно,
   но не совпадают полностью, что связано с различием их альбедо
   и влиянием на температуру среды.

4. Несмотря на колебания, популяции не исчезают и не растут
   неограниченно, что свидетельствует о наличии механизма
   саморегуляции системы.

Таким образом, модель Daisyworld демонстрирует устойчивую
динамику с колебаниями численности двух видов ромашек.
